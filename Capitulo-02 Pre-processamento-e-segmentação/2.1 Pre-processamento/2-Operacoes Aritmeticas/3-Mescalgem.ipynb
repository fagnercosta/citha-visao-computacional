```python
import cv2 as cv
from matplotlib import pyplot as plt
		
carro = cv.imread('carro.jpg', cv.IMREAD_COLOR)
estrada = cv.imread('estrada.jpg', cv.IMREAD_COLOR)
tamanhoCarro = carro.shape
estrada = cv.resize(estrada, (tamanhoCarro[1], tamanhoCarro[0]))
		
		
imgMesclada = cv.addWeighted(carro, 0.5, estrada, 0.5, 0)
		
fig=plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(131)
plt.imshow(carro)
plt.title('Carro')
		
ax2 = fig.add_subplot(132)
plt.imshow(estrada)
plt.title('Estrada')
		
ax3 = fig.add_subplot(133)
plt.imshow(imgMesclada)
plt.title('Resultado')		
plt.show()
```
