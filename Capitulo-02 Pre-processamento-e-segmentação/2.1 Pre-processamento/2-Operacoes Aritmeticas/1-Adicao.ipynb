```python
import cv2 as cv
from matplotlib import pyplot as plt
		
img1 = cv.imread('fichas-posicao1.png',1)
img2 = cv.imread('fichas-posicao2.png',1)
tamanhoImg1 = img1.shape
print(tamanhoImg1)
# (largura, altura)
img2 = cv.resize(img2, (tamanhoImg1[1], tamanhoImg1[0])) 
imgSomada = cv.add(img1, img2)
#Plotando as imagens com Matplotlib
grafico = plt.figure(figsize=(20,5))
ax1 = grafico.add_subplot(131)
plt.imshow(cv.cvtColor(img1, cv.COLOR_BGR2RGB))
plt.title('Imagem 01')
		
ax2 = grafico.add_subplot(132)
plt.title('Imagem 02')
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
		
ax3 = grafico.add_subplot(133)
plt.title('Somada RGB')
plt.imshow(cv.cvtColor(imgSomada, cv.COLOR_BGR2RGB))
		
plt.show()
```
