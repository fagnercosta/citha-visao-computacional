```python
import cv2
from matplotlib import pyplot as plt

imagemOriginal = cv2.imread('rodasFalhas.png')
elemento_estruturante = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (7, 7))
imagemProcessada = cv2.morphologyEx(imagemOriginal, cv2.MORPH_CLOSE, elemento_estruturante)

plt.figure(figsize=(10, 8))
plt.subplot(1, 2, 1)
plt.imshow(imagemOriginal)
plt.title("Imagem Original com Falhas")

plt.subplot(1, 2, 2)
plt.imshow(imagemProcessada)
plt.title("Imagem Tratada com Fechamento")

plt.show()
```
