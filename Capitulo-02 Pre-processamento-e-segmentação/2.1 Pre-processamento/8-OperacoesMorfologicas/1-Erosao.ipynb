```python
import cv2
from matplotlib import pyplot as plt

imagemOriginal = cv2.imread("letra-J-com-ruidos.png")
elemento_estruturante = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
imagemProcessada = cv2.erode(imagemOriginal, elemento_estruturante, iterations=1)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(imagemOriginal)
plt.title("Imagem Original")

plt.subplot(1, 2, 2)
plt.imshow(imagemProcessada)
plt.title("Imagem com Erosão")

plt.show()
```
