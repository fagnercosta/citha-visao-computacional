```python
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline

imgemOriginal = cv.imread('head.png')
grauSuavizacao = 3

imgFiltrada = cv.medianBlur(imgemOriginal, grauSuavizacao)

fig = plt.figure(figsize=(20,50))

ax1 = fig.add_subplot(121)
plt.imshow(imgemOriginal)

plt.title("Imagem original com rudo")
ax2 = fig.add_subplot(122)

plt.imshow(imgFiltrada)
plt.title("Imagem com filtro Mediana")

plt.show()
```
