```python
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('001.png')
#Mascara 
mascara = (5,5)
#Grau de suavização
grauSuavizacao = 0
imgGauss = cv.GaussianBlur(img, mascara, grauSuavizacao)

fig = plt.figure(figsize=(20,50))
ax1 = fig.add_subplot(121)

plt.imshow(img)
plt.title("Imagem original com ruído")

ax2 = fig.add_subplot(122)
plt.imshow(imgGauss)

plt.title("Imagem Filtro Gaussiano")
plt.show()
```
