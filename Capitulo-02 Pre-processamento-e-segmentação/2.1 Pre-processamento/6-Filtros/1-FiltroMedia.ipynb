```python
import cv2 as cv
from matplotlib import pyplot as plt

imgOriginal = cv.imread('001.png')

#Criar a mascara
mask = (5,5)
#aplica filtro mdia na imagem
imgFiltrata = cv.blur(imgOriginal, mask)

fig = plt.figure(figsize=(20,40))
ax1 = fig.add_subplot(121)
#plt.imshow(img)
plt.imshow(cv.cvtColor(imgOriginal, cv.COLOR_BGR2RGB))
plt.title("Imagem original com rudo")

ax2 = fig.add_subplot(122)
plt.imshow(cv.cvtColor(imgFiltrata, cv.COLOR_BGR2RGB))
plt.title("Imagem Filtrada")
```
