```python
import cv2
import numpy as np
from datetime import datetime



cap = cv2.VideoCapture(0)


# Subtrair Fundo
subtractor = cv2.createBackgroundSubtractorMOG2(history=100, varThreshold=50, detectShadows=True)


while True:
    
 
    ret, frame = cap.read()
    
    if ret == True:
        
             # Aplicar Filtro Gaussuano
            frame = cv2.GaussianBlur(frame, (5, 5), 0)

            # Criar Mascara sem o fundo..
            mask = subtractor.apply(frame)
            # Binarizar para aplicar operações morfologicas...
            ret, mask = cv2.threshold(mask,90,255,cv2.THRESH_BINARY)

            # Criar Elemento Estrutura..
            kernel = np.ones((2,2),np.uint8)

            # Realizar Abertura 
            mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel, iterations = 2)

            # Obter Contornos
            contours, hierarchy = cv2.findContours(mask,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

            # Obter Lista de contornos
            for cnt in contours:

                # Obter Área
                area = cv2.contourArea(cnt)

                if area > 2000:

                    # Obter coordenadas da área
                    x,y,w,h = cv2.boundingRect(cnt)


                    # Gerar nome do arquivo com data e hora
                    data_hora = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")  # Formato: 2025-07-07_14-30-15
                    nome_arquivo = f"objeto_detectado_{data_hora}.jpg"

                    # Salvar a imagem
                    cv2.imwrite(nome_arquivo, frame)


            cv2.imshow("Frame", frame)
            cv2.imshow("mask", mask)



            if cv2.waitKey(10) & 0xFF == ord('q'):
                break
        
    else:
         print('Vídeo Finalizou..')
       
        
cap.release()
cv2.destroyAllWindows()
```
