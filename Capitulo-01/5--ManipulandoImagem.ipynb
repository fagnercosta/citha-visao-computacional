```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

ImagemMatriz = Image.open("./Imagens/cubo.png")
ImagemMatriz = np.asarray(ImagemMatriz, dtype=np.float32)/255

# zera camada vermelha
ImagemMatriz[:,:,1] = 0
# zera camada verde
ImagemMatriz[:,:,2] = 0

# visualizando
plt.figure(figsize=(10, 10))
im = plt.imshow(ImagemMatriz, aspect='auto')
plt.axis("off")
plt.show()
```
