```python
from ultralytics import YOLO
import cv2

# Carregar o modelo pré-treinado YOLOv8n
modelo = YOLO("yolov8n.pt")

# Iniciar a captura da webcam
cap = cv2.VideoCapture(0) # "0" é a webcam padrão

if not cap.isOpened():
    print("Erro: Não foi possível abrir a webcam")
    exit()

print("Webcam iniciada. Pressione 'q' para fechar.")

while True:
    # Capturar frame da webcam
    ret, frame = cap.read()
    if not ret:
        print("Erro: Falha ao capturar o frame")
        break
    
    # Fazer a detecção no frame com o modelo YOLO
    results = modelo.predict(source=frame, show=False, conf=0.5)
    
    # O método .plot() desenha as detecções diretamente no frame
    frame_com_deteccoes = results[0].plot()
    
    # Exibir o frame com as detecções
    cv2.imshow("YOLOv8 - Detecção via Webcam", frame_com_deteccoes)
    
    # Fechar a janela ao pressionar a tecla 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("Fechando a câmera...")
        break

# Liberar a webcam e fechar todas as janelas do OpenCV
cap.release()
cv2.destroyAllWindows()
```
